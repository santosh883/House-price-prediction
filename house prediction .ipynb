```python
#define the inpute data (x) and define the output/prediction(y)
y=db['price']
x=db.drop(['price','date'],axis=1)
#but we need to drop some colums lite price as it is alredy in y and some useless coiums
```

```python
#let us train the data on linearRegression model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#now we need to split the data into training set and testing set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
